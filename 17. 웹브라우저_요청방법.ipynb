```python
import requests
```

```python
url ='https://httpbin.org/get'
response = requests.get(url)
response.status_code
```

```python
print(response.text)
```

get
- 서버에 데이터를 요청해서 가져옴.
- URL에 파라미터가 붙음.
- 데이터를 바꾸지 않음 (안전함)

```python
params = {'name':'daeun',
          'age':20}

url ='https://httpbin.org/get'
response = requests.get(url,params=params)
response.status_code
print(response.text)
```

post
- 서버에 데이터를 전송해서 새로 만듬.
- DB나 상태가 바뀜

```python
url = 'https://httpbin.org/post'
response = requests.post(url, data = {'key':'value'})
print(response.text)
```

```python
response.json()['headers']['Content-Type']
```

```python
response.json()['headers']['User-Agent']
```

put
- 서버에 기존의 리소스를 통째로 교체
- 지정한 리소스를 완전히 덮어씀
- 일반적으로 post보다 '수정'의 의미가 더 강함

```python
response = requests.put('https://httpbin.org/put', data={'key':'value'})
print(response.text)
```

patch
- 리소스의 일부만 수정
- put이랑 비슷한데, 일부 필드만 바꿈
- rest api 설계에서 자주 씀

```python
response = requests.patch('https://httpbin.org/patch', data={'key': 'value'})
print(response.text)
```
