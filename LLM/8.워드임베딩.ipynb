# 단어의 임베딩
 - 빈도수 계산 : 빈도기반 (TF, 상대빈도)
 - TDM : matrix, TF를 행렬로 만드는 것, 사전을 이용한 단순 빈도
 - TF-IDF : TF * IDF
 - IDF: 역 문서 빈도

```python
text = 'John likes to watch movies. Many likes movies too.\
    Mary also likes to watch football games'
    
words = text.replace('.','').split()
print(words)
```

```python
# 딕셔너리 TF생성
import numpy as np
word_count = np.unique(words, return_counts = True)
word_to_cnt={}
for word , cnt in zip(*word_count):
    word_to_cnt[word] = cnt
print(word_count)
```

TDM (문장 별로 단어의 빈도수를 계산해서 행렬로 만들어 놓은 것)

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['John likes to watch movies. Many likes movies too.',
    'Mary also likes to watch football games']
    
vector = CountVectorizer()
tdm_array = vector.fit_transform(corpus).toarray()
tf_dic = vector.vocabulary_
print(tdm_array)
print(tf_dic)
```

```python
import pandas as pd
tf_dic_sorted = dict(sorted(tf_dic.items(), key = lambda item: item[1]))
tdm = pd.DataFrame(tdm_array, columns = tf_dic_sorted.keys())
print(tdm)
```

```python
# TF_IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec = TfidfVectorizer()
tfidf_array = tfidf_vec.fit_transform(corpus).toarray()
tfidf_dic = tfidf_vec.vocabulary_
tfidf_dic_sorted = dict(sorted(tfidf_dic.items(), key = lambda item:item[1]))
tfidf_dtm = pd.DataFrame(tfidf_array, columns = tfidf_dic_sorted.keys())
print(tfidf_dtm)
```

```python
from gensim.models import Word2Vec
word_list = []

corpus = ['John likes to watch movies. Many likes movies too.',
          'Mary also likes to watch football games']

for word in corpus:
    word_list.append(word.replace('.','').split())
    
word_list
```

```python
model = Word2Vec(word_list, sg = 0, vector_size = 100, window = 3, min_count = 1)
print(model.wv.most_similar('likes'))
print(model.wv.similarity('movies','games'))
```

imdb 영화 리뷰데이터 실습

```python
# tensorflow
from tensorflow.keras.datasets import imdb
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words = 10000)
(X_train, y_train), (X_test, y_test)
```

```python
# 텍스트 데이터의 전처리
from tensorflow.keras.preprocessing.sequence import pad_sequences
# 영화평안의 사이즈를 맞추기 위해서 패딩을 사용
X_train_pad = pad_sequences(X_train, maxlen = 80, padding = 'post', truncating = 'post')
X_test_pad = pad_sequences(X_test, maxlen = 80, padding = 'post', truncating = 'post') 
```

```python
from tensorflow.keras import Sequential, layers
model_dnn = Sequential([
    layers.Input(shape = (80,)), # 80개의 단어를 입력 
    layers.Embedding(input_dim = 10000, output_dim = 32), # (80,32)
    layers.Flatten(),
    layers.Dense(64, activation = 'relu'),
    layers.Dense(2, activation = 'softmax')]
)
```

```python
model_dnn.summary()
```

```python
model_dnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_dnn.fit(X_train_pad, y_train, epochs=10, batch_size=200)
```

```python
model_dnn.evaluate(X_test_pad, y_test)
```

RNN model

```python
model_rnn = Sequential([
    layers.Input(shape = (80,)), # 80개의 단어를 입력 
    layers.Embedding(input_dim = 10000, output_dim = 32), # (80,32)
    #layers.Flatten(),
    #layers.Dense(64, activation = 'relu'),
    layers.SimpleRNN(64),
    layers.Dense(2, activation = 'softmax')] # 출력: 긍정, 부정
)
```

```python
model_rnn.summary()
```

```python
model_rnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_rnn.fit(X_train_pad, y_train, epochs = 10, batch_size= 200)
model_rnn.evaluate(X_test_pad, y_test)
```

RNN 의 구조의 한계
- 단기기억이 안좋다는 단점을 가지고 있어서 다른 모델이 나옴.

### 모델의 정확도 높여보기

```python
# 모델의 전처리
X_train_pad = pad_sequences(X_train, maxlen = 100, padding = 'pre', truncating = 'pre')
X_test_pad = pad_sequences(X_test, maxlen = 100, padding = 'pre', truncating = 'pre')

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Input
# 모델 생성하기
model_rnn_4 = Sequential([
    Input(shape=(80,)), 
    Embedding(input_dim=10000, output_dim=32), 
    SimpleRNN(64, return_sequences=True), # 다음 레이어로 출력을 전달
    SimpleRNN(128), # return_sequences=True가 없으므로 마지막 출력만 전달
    Dense(2, activation='softmax')
])

model_rnn_4.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_rnn_4.fit(X_train_pad, y_train, epochs=50, batch_size=128)
model_rnn_4.evaluate(X_test_pad, y_test)
```

**손실률(Loss)**은 모델이 틀린 데이터를 기준으로 정답과 얼마나 멀게 예측했는지를 나타내는 지표입니다. 📉

손실은 모델이 얼마나 자신감 있게 틀렸는지를 측정한다고 생각하면 됩니다.

그럼 나의 모델은 자신감있게 틀렸다고 보면 될 것 같다.....

RNN은 마지막의 결과에 더 포커스를 맞추는 경향이 있다. 그래서 글자의 수를 줄이면 정확도가 놓아질 수 있음

# CNN model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalAveragePooling1D, Dense, Input

X_train_pad = pad_sequences(X_train, maxlen = 100, padding = 'pre', truncating = 'pre')
X_test_pad = pad_sequences(X_test, maxlen = 100, padding = 'pre', truncating = 'pre')

model_cnn = Sequential([
    Input(shape=(100,)), 
    Embedding(input_dim=10000, output_dim=32),
    
    Conv1D(64, 3, activation='relu'),
    GlobalAveragePooling1D(),
    
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid') # 긍정/부정 이진 분류를 위한 출력 레이어
])

model_cnn.summary()
```

```python
model_cnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_cnn.fit(X_train_pad, y_train, epochs = 10, batch_size= 200)
model_cnn.evaluate(X_test_pad, y_test)
```

RNN(Recurrent Neural Network)이 문장을 순서대로 읽는 것과 달리, CNN은 중요한 단어 조합을 빠르게 파악할 수 있다는 장점이 있습니다. '정말 별로다'라는 구문이 문장의 맨 앞에 있든, 중간에 있든, 끝에 있든 상관없이 CNN은 이 패턴을 즉시 인식할 수 있습니다.

따라서 CNN은 텍스트 분류에서 문맥의 순서보다는 단어 간의 조합이 중요할 때 RNN보다 더 빠르고 효율적인 성능을 보여주기도 합니다.

⭐ 실제로 정확도는 높아지고 손살율도 많이 줄어든 것을 볼 수 있다.
