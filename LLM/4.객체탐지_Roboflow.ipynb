```python
import os
from dotenv import load_dotenv
loaded = load_dotenv()
```

```python


from inference_sdk import InferenceHTTPClient
CLIENT = InferenceHTTPClient(
    api_url = 'https://detect.roboflow.com',
    api_key = os.getenv("ROBOFLOW_API_KEY")
)
```

```python
from PIL import Image
import numpy as np

img = Image.open('가위바위보.jpg')     # 이미지 열기
img_array = np.array(img) 
img    
```

```python
result = CLIENT.infer(img,model_id='rock-paper-scissors-sxsw/14')
result
```

```python
import cv2
img = cv2.imread('1.jpg')



for pred in result['predictions']:
  x = pred['x']
  y = pred['y']
  width = pred['width']
  height = pred['height']
  conf = pred['confidence']
  obj_class = pred['class']
  
  x1, y1 = int(x-width/2), int(y-height/2)
  x2, y2 = int(x+width/2), int(y+height/2)
  cv2.rectangle(img, (x1,y1), (x2, y2), (0,0,255), 2)
  cv2.putText(img, f'{obj_class} {conf:.4f}', (x1,y1), cv2.FONT_HERSHEY_PLAIN, 2, (0,0,255))
    
  print(x,y,width,height,conf,obj_class)
  
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyWindow()
```

```python
from roboflow import Roboflow
import cv2
rf = Roboflow(api_key=os.getenv('ROBOFLOW_API_KEY'))
project = rf.workspace().project('numbers-yt0h2')
numbers_model = project.version(2).model
numbers_model
```

```python
img_path = "test.jpg"
predictions = numbers_model.predict(img_path, confidence=40, overlap=30).json()

print(predictions)
```

```python
import cv2
from matplotlib import pyplot as plt

# 이미지 불러오기
img_path = "test.jpg"
img = cv2.imread(img_path)

# OpenCV는 BGR, Matplotlib는 RGB이므로 변환
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# predictions는 Roboflow 모델의 JSON 결과
for pred in predictions['predictions']:
    x, y = pred['x'], pred['y']
    width, height = pred['width'], pred['height']
    conf = pred['confidence']
    obj_class = pred['class']

    # 좌상단(x1,y1), 우하단(x2,y2) 계산
    x1, y1 = int(x - width/2), int(y - height/2)
    x2, y2 = int(x + width/2), int(y + height/2)

    # 사각형 그리기
    cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (255, 0, 0), 2)
    # 클래스명과 신뢰도 표시
    cv2.putText(img_rgb, f"{obj_class} {conf:.2f}", (x1, y1 - 5),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

# Jupyter Notebook에서 이미지 표시
plt.figure(figsize=(8,8))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
