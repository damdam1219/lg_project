```python
import requests
import folium
from pandas import json_normalize
```

```python
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,
'ins_lng': 126.9863309,
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 600,
'set_date': ''
})
starbucks=req.json()
print(type(starbucks))
```

```python
st_df = json_normalize(starbucks, 'list')
st_df
```

```python
st_df_map = st_df[['s_name','sido_code','sido_name','gugun_code','gugun_name',
'doro_address','lat','lot']]
st_df_map.info()
```

```python
st_df_map['lat'] = st_df_map['lat'].astype(float)
st_df_map['lot'] = st_df_map['lot'].astype(float)
st_df_map.head()
```

```python
loc=[37.5662952, 126.9779451] #위도,경도
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,
'ins_lng': 126.9863309,
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 600,
'set_date': ''
})
starbucks=req.json()
print(type(starbucks))

['s_name','sido_code','sido_name','gugun_code','gugun_name',
'doro_address','lat','lot']
```

```python
import folium

# 서울 중심 좌표로 지도 생성
m = folium.Map(location=loc, zoom_start=12)

# 스타벅스 매장 리스트: 'list' 키가 있다고 가정 (API 구조에 따라 다름)
for store in starbucks['list']:  
    name = store['s_name']
    lat = float(store['lat'])
    lon = float(store['lot'])
    
    folium.Marker(
        location=[lat, lon],
        popup=name,
        icon=folium.Icon(color='green', icon='coffee', prefix='fa')
    ).add_to(m)

# 지도 출력 (Jupyter 기준)
m
```
