```python
# 날짜와 가격 데이터
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.title("주식가격 5일")
plt.plot(dates, prices, marker='o', label='Stock Price')  # plt.plot() 사용, marker 옵션 추가
plt.xlabel("날짜")
plt.ylabel("가격")
plt.ylim(96,110)
plt.legend()   # legend() + label='문자열'로 지정 → ✅ 
plt.grid(True)
plt.show()
```

```python
import seaborn as sns 
tips_df = sns.load_dataset('tips')
tips_df
```

```python
tips_df['size'].value_counts()
```

```python
# 히스토그램

fig = plt.figure(figsize=(12,5))
a1 = fig.add_subplot(1,2,1) # total_bill
a2 = fig.add_subplot(1,2,2) # tip

a1.hist(data = tips_df, x='total_bill',bins=20)
a2.hist(data = tips_df, x='tip',bins=20)

plt.tight_layout()
plt.show()
```

```python
# boxplot

fig = plt.figure(figsize=(12,5))
a1 = fig.add_subplot(1,1,1) # total_bill

a1.boxplot(data = tips_df, x= 'total_bill') # 일변량 차트
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)  #total_bill
axes.boxplot(
  x=[
    tips_df[tips_df['time']=='Female']['tip'],
    tips_df[tips_df['time']=='Male']['tip']
  ],
  labels=['Female','Male']
)
plt.show()
    
    
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1)

axes.boxplot(
  x=[
    tips_df[(tips_df['smoker'] == 'Yes') & (tips_df['sex'] == 'Male')]['tip'],
    tips_df[(tips_df['smoker'] == 'No') & (tips_df['sex'] == 'Male')]['tip'],
    tips_df[(tips_df['smoker'] == 'Yes') & (tips_df['sex'] == 'Female')]['tip'],
    tips_df[(tips_df['smoker'] == 'No') & (tips_df['sex'] == 'Female')]['tip']
  ],
  labels=['Smoker (Male)', 'Non-Smoker (Male)','hh','hh']
)

plt.show()
```

```python
# 산점도
colors = {"Female":"red","Male":"blue"}
tips_df['tip_qct'] = tips_df['tip']/tips_df['total_bill']
fig = plt.figure()
a1 = fig.add_subplot(1,1,1)

a1.scatter(data=tips_df, x='total_bill',y='tip',s=tips_df['tip_qct']*1000,
           c=tips_df['sex'].map(colors))
fig.show()
```

```python
# 파이차트 
# 성별의 비
# 빈도수
sex_count = tips_df['sex'].value_counts()
smoker_count = tips_df['smoker'].value_counts()

fig = plt.figure()
a1 = fig.add_subplot(1,1,1)

a1.pie(smoker_count, labels=smoker_count.index, colors=['blue', 'red'], autopct='%1.1f%%')
plt.show()
```
