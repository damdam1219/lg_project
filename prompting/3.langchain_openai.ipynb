# async, sync

| 구분    | 동기(sync)     | 비동기(async)               |
| ----- | ------------ | ------------------------ |
| 실행 방식 | 작업 완료까지 대기   | 작업 대기 중 다른 작업 수행 가능      |
| 코드 작성 | 간단, 직관적      | 복잡할 수 있으나 효율적            |
| 사용 예  | 계산, 간단한 입출력  | 네트워크 요청, 파일 입출력, 병렬 처리 등 |
| 장점    | 이해 및 디버깅 쉬움  | 대기 시간 절약, 효율적 자원 활용      |
| 단점    | 느리고 자원 낭비 가능 | 코드 복잡도 증가 가능             |


```python
import asyncio

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)

async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
```

```python
import time
messages = [
    ('system', '당신은 서울 동작구 노량진의 음식과 문화 전문가입니다.'),
    ('human', '노량진을 대표하는 음식을 맛 볼 수 있는 식당 5개를 추천해주세요')
]

start = time.perf_counter()
await invoke_parallel(llm, messages)
end = time.perf_counter()

print(f'Elapsed time:{end-start:.2f} second')
```

```python
# sync 호출 시간 비교 5 time
import asyncio

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)

async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
    
import time
messages = [
    ('system', '당신은 서울 동작구 노량진의 음식과 문화 전문가입니다.'),
    ('human', '노량진을 대표하는 음식을 맛 볼 수 있는 식당 5개를 추천해주세요')
]

start = time.perf_counter()
await invoke_async(llm, messages)
end = time.perf_counter()

print(f'Elapsed time:{end-start:.2f} second')
```
