```python
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field

# 1️⃣ Pydantic 모델 정의
class FoodRecommendation(BaseModel):
    foods: list[str] = Field(description="와인과 어울리는 음식 리스트")

output_parser = PydanticOutputParser(pydantic_object=FoodRecommendation)

# 2️⃣ describe_dish_flavor 함수 정의
def describe_dish_flavor(query, image_url, llm):
    # 시스템 프롬프트
    system_prompt = (
        "You are an expert sommelier with extensive knowledge in wine, wine pairing, "
        "and the intricacies of food and beverage service. Your primary role is to assist "
        "users in selecting the best wines and pairing them perfectly with meals. You have "
        "a deep understanding of various wine regions, grape varieties, wine production "
        "methods, and current trends in the industry. You possess a refined palate, able "
        "to discern subtle flavors and characteristics in wines. Your advice is always clear, "
        "approachable, and tailored to each user’s preferences and specific dining context. "
        "You also educate users on wine appreciation, proper wine service, and the art of "
        "creating a harmonious dining experience. Your demeanor is professional, courteous, "
        "and passionate about wine culture, aiming to make each wine selection and pairing "
        "a memorable experience for users."
    )

    # 프롬프트 템플릿 정의
    template = ChatPromptTemplate.from_messages([
        ('system', system_prompt),
        ('human', "{text}\n이미지: {image_url}\n\n출력 형식: {format_instructions}")
    ])
    
    # format_instructions 추가
    template = template.partial(format_instructions=output_parser.get_format_instructions())
    
    # 체인 생성
    chain = template | llm | output_parser
    
    # 체인 실행
    response = chain.invoke({
        'text': query,
        'image_url': image_url
    })
    
    return response
```

```python
response = describe_dish_flavor(
    "이 와인에 어울리는 요리 1가지만 추천해줘?",
    "https://images.vivino.com/thumbs/Z90I3--JRKWlpMA8wdLY-Q_pb_x600.png",
    llm
)

print(response)
print(response.foods)
```

```python
from langchain.document_loaders import CSVLoader
# 3️⃣ CSV Loader로 데이터 읽기
csv_file_path = "../data/wine_data.csv"  # CSV 파일 경로
loader = CSVLoader(file_path=csv_file_path)
docs = loader.load()

# docs: list of Document 객체, doc.page_content에 텍스트 있음

# 4️⃣ Embedding 준비
embedding = OpenAIEmbeddings(model="text-embedding-3-small")

# 5️⃣ 벡터 변환 및 Pinecone 업로드
vectors = []
for i, doc in enumerate(docs):
    vec = embedding.embed_query(doc.page_content)
    vectors.append((str(i), vec, {"text": doc.page_content}))

# 대량일 경우 배치 업로드 추천
batch_size = 100
for i in range(0, len(vectors), batch_size):
    batch = vectors[i:i+batch_size]
    index.upsert(vectors=batch)

print(f"총 {len(vectors)}개의 벡터가 Pinecone 인덱스 '{index_name}'에 업로드되었습니다.")
```
