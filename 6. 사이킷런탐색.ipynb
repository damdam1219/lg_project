 1️⃣ 이름을 해부해보자
Stratified 👉 계층화 샘플링

각 fold가 **타깃 클래스 비율(레이블 분포)**을 유지하도록 나눔

예: 타깃이 1:9 (양성:음성)이라면, 각 fold에도 비슷한 비율로 유지

Group 👉 그룹 단위로 샘플링

같은 그룹에 속한 샘플은 무조건 같은 fold에 묶음

예: 같은 사람(환자)에서 나온 여러 데이터, 같은 회사에서 나온 데이터 등 그룹화된 데이터를 떨어뜨리지 않음

KFold 👉 k-폴드 교차검증

📌 2️⃣ 언제 쓰나?
둘 다 동시에 필요한 상황!

예: 한 사람의 여러 사진이 데이터에 있고, 분류 문제인데 클래스 불균형이 있음.

같은 사람의 사진이 학습셋과 테스트셋에 동시에 들어가면 과적합 위험.

클래스 불균형도 유지해야 정확한 평가 가능.

그래서 그룹 단위로 묶고, 클래스 비율도 맞추는 게 StratifiedGroupKFold!

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns=iris.feature_names )
iris_df['label']  = iris.target
iris_df['label'].value_counts()
```

```python
iris_df
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris_df, iris_df['label']):
  # train_index: 학습용 데이터로 사용할 샘플의 인덱스/ test_index: 검증용데이터로 사용할 샘플의 인덱스
  y_train = iris_df['label'].iloc[train_index]
  y_test =  iris_df['label'].iloc[test_index]
  print(f'train 분포: {y_train.value_counts()}')
  print(f'test 분포: {y_test.value_counts()}')
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris_df, iris_df['label']):
    train_data = iris_df.iloc[train_index]
    test_data = iris_df.iloc[test_index]
    X_train = train_data[iris.feature_names]
    X_test = test_data[iris.feature_names]

    y_train = train_data['label']
    y_test =  test_data['label']

    model = DecisionTreeClassifier(random_state=156)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    acc = accuracy_score(y_test, pred)
    print(acc)
    
```

와인 데이터 스케일 전

```python
wine = load_wine()
X= wine.data
y= wine.target

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=11)

model.fit(X_train, y_train)
pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

와인 데이터 스케일 후

```python
X= wine.data
y= wine.target

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000) # 수렴을 위한 하이퍼파라미터

from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=11)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

model.fit(X_train_scaled, y_train)
pred = model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, pred)
print(acc)
```

knn 모델과 ranfomforst모델의 스케일 전/후 데이터 모델 성능

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# 스케일 전 데이터
# KNeighborsClassifier
model_n = KNeighborsClassifier(n_neighbors=5) # 기본디폴트: 5

model_n.fit(X_train, y_train)

pred_n = model_n.predict(X_test)

acc_n = accuracy_score(y_test, pred_n)
print(f'KNN 모델 사용 정확도 : {acc_n:.4%}')

# RandomForestClassifier
model_k = RandomForestClassifier()

model_k.fit(X_train, y_train)

pred_k = model_k.predict(X_test)

acc_k = accuracy_score(y_test, pred_k)
print(f'랜덤포레스트 모델 사용 정확도 :{acc_k:.4%}')
```

```python
# 스케일 후 데이터
# KNeighborsClassifier
model_n = KNeighborsClassifier()

model_n.fit(X_train_scaled, y_train)

pred_n = model_n.predict(X_test_scaled)

acc_n = accuracy_score(y_test, pred_n)
print(f'KNN 모델 사용 정확도 : {acc_n:.4%}')

# RandomForestClassifier
model_k = RandomForestClassifier()

model_k.fit(X_train_scaled, y_train)

pred_k = model_k.predict(X_test_scaled)

acc_k = accuracy_score(y_test, pred_k)
print(f'랜덤포레스트 모델 사용 정확도 :{acc_k:.4%}')
```
