## scatter
상관관계 시각화 -  분포, 경향성, 이상치 탐색
- 회귀분석, 군집분석 등에서 탐색적 데이터 분석에 활용

```python
sns.scatterplot(x=  df['total_bill'],y= df['tip'],hue=df['sex'],size = df['size'],palette='pastel')
```

- size: 점의 크기를 지정하는 변수(수치형 변수로, 점의 크기를 조정) 
-  hue: 추가적인 범주형 변수를 사용하여 점의 색상을 구분 
-  style: 점의 스타일을 구분하는 변수 
-  palette: 색상 팔레트를 지정하여 다양한 색상을 사용할 수 있음(예: 'deep', 'pastel')

# boxplot

- 통계량 확인 가능
- 분포 확인

```python
car_df = pd.read_csv('data/EV_charge.csv')
car_df
```

```python
sns.boxplot(x= 'weekday', y = 'kwhTotal',data = car_df)
```

원본 데이터 + 통계량을 같이 출력하여 표시한다. -> boxplot + stripplot

```python
sns.boxplot(x= 'weekday', y = 'kwhTotal',data = car_df)
sns.stripplot(x= 'weekday', y = 'kwhTotal',data = car_df)
```

# swarmplot
- 값이 겹치지 않는다.
- 분포를 잘 보여준다.
- 단점: 성능 저하 고려

```python
sns.boxplot(x= 'weekday', y = 'kwhTotal',data = car_df)
sns.stripplot(x= 'weekday', y = 'kwhTotal',data = car_df)
sns.swarmplot(x= 'weekday', y = 'kwhTotal',data = car_df,color='grey',alpha = 0.4)
```

```python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sns.boxplot(x= 'weekday', y = 'kwhTotal',data = car_df , order = weekday_order)
```

```python
sns.boxplot(x= 'weekday', y = 'kwhTotal',data = car_df ,hue ='platform', order = weekday_order)
```

```python
# plotly + boxplot
weekday_order = {
    'weekday': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
} # 딕셔너리 형태로 들어가야함.

px.box(data_frame=car_df, x='weekday',y = 'kwhTotal',width=400,height=400, points ='all'
       , category_orders=weekday_order)
```

```python
px.box(data_frame=car_df, x='weekday',y = 'kwhTotal',width=400,height=400, points ='all'
       , category_orders=weekday_order,color='platform')
```

# heatmap
- 데이터를 요약(합계, 평균,,, ) 한 결과를 샥깔로 시각화하는 방법
- 요약: 피봇 테이블 (탐색) -> 인사이트 시각화로 히트맵을 작성

```python
# 요약
pivot_df = tip_df.pivot_table('tip', 'day','size')
sns.heatmap(pivot_df, annot=True) # 실제 값을 넣어줌
```

⭐null처리를 해줘야 함.

```python
pivot_df.fillna(0, inplace=True)
```

```python
sns.heatmap(pivot_df, annot=True)
```

```python
# plotly.imshow()
px.imshow(
    pivot_df, x = pivot_df.columns, y = pivot_df.index.astype('str')
                                ,width = 400, height = 400, text_auto = '.2e'
)
```
